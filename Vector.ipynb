```python
import time
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
n = int(input("Enter the size of vector -> "))
minimum = int(input("Enter minimum number for random -> "))
maximum = int(input("Enter maximum number for random -> "))
```

### Serial

```python
def serial_random_array(n, minimum, maximum):
    vector = []
    for i in range(n):
        temp = []
        for j in range(n):
            temp.append(random.randint(minimum, maximum))
        vector.append(temp)
    return vector
```

```python
def serial_vector_multiply(a, b, n):
    c = []
    for i in range(n):
        temp = []
        for j in range(n):
            temp.append(a[i][j] * b[i][j])
        c.append(temp)
    return c    
```

```python
def serial(n,minimum, maximum):
    st = time.time()
    a = serial_random_array(n, minimum, maximum)
    b = serial_random_array(n, minimum, maximum)
    c = serial_vector_multiply(a, b, n)
    return time.time() - st
```

### Parallel

```python
def parallel_random_vector(n, minimum, maximum):
    return np.random.randint(minimum, maximum, (n,n))
```

```python
def parallel_vector_multiply(np_a, np_b):
    return np_a * np_b
```

```python
def parallel(n, minimum, maximum):
    st = time.time()
    np_a = parallel_random_vector(n, minimum, maximum)
    np_b = parallel_random_vector(n, minimum, maximum)
    np_c = parallel_vector_multiply(np_a, np_b)
    return time.time() - st
```

```python
print("Serial Time :", serial(n, minimum, maximum))
```

```python
print("Parallel Time :", parallel(n, minimum, maximum))
```

### Visualization

```python
x = []
s = []
p = []
for k in range(1,200):
    x.append(k)
    s.append(serial(k, minimum, maximum))
    p.append(parallel(k, minimum, maximum))
```

```python
plt.figure(figsize = (20,10), dpi = 500)
plt.plot(x, s, 'r', label = 'serial')
plt.plot(x, p, 'b', label = 'parallel')
plt.title('Serial VS Parallel graph')
plt.xlabel('size of vector')
plt.ylabel('time')
plt.legend()
plt.show()
```
